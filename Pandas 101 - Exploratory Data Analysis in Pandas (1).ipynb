#####  Countries with the highest percentage population in the world 

```python
df1=df[['Country','Continent','World Population Percentage']].sort_values(by='World Population Percentage',ascending=False).reset_index().head()
plt.pyplot.bar(df1['Country'].unique().tolist(),df1['World Population Percentage'])
```

##### Continent vs World Population Percentage

```python
df1=df[['Country','Continent','World Population Percentage']].groupby(by='Continent',as_index=True).sum().sort_values(by='World Population Percentage',ascending=False)
df1.reset_index(inplace=True)
plt.pyplot.bar(df1['Continent'].unique().tolist(),df1['World Population Percentage'])
```
