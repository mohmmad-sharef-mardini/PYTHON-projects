## DATA CLEANING 

```python

#replacing ? with nan value

df.replace('?',np.nan,inplace=True)
```

```python
# handle with missing value

missing_data=df.isnull()

for col in df.columns.values.tolist() :
    print(missing_data[col].value_counts())
    
```

```python
#Replacing nan value 
#Replacing  normalized-losses and stroke and horsepower and peak-rpm with avarage 

ava_loss=df["normalized-losses"].astype("float").mean()
df.replace(np.nan,ava_loss,inplace=True)

ava_stroke=df["stroke"].astype("float").mean()
df.replace(np.nan,ava_stroke,inplace=True)

ava_horsepower=df["horsepower"].astype("float").mean()
df.replace(np.nan,ava_horsepower,inplace=True)

ava_peak=df["peak-rpm"].astype("float").mean()
df.replace(np.nan,ava_peak,inplace=True)
```

```python
#Replacing  num-of-doors  with mode

freq_doors=df['num-of-doors'].value_counts().idxmax()
#another way to get mode  
freq_doors=df['num-of-doors'].mode()

df.replace(np.nan,freq_doors,inplace=True)

```

```python
#Replacing  bore with median  

median_bore=df['bore'].astype('float').median()
df.replace(np.nan,median_bore,inplace=True)
```

```python
#Remove row that is nan value in price column

df.dropna(subset=['price'],axis=0,inplace=True)

#Reset Index

df.reset_index(drop=True,inplace=True)

```

```python
# cheak out the data if the data has a missing value 
missing_data=df.isnull()

for col in df.columns.values.tolist() :
    print(missing_data[col].value_counts())
```

```python
# now we have data without missing value
```

### data formating


```python
#convert type of data

df[["bore", "stroke","peak-rpm","price","horsepower"]]=df[["bore", "stroke","peak-rpm","price","horsepower"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
```

### data standerization 

```python
 

df['city-L/100km'] = 235/df["city-mpg"]
df["highway-L/100km"]=235/df["highway-mpg"]
```

### Data normalization

```python
#data normalization

# simple feature scaling - min max - zscore

df['length'] = df['length']/df['length'].max()
df['width'] = (df['width']-df['width'].min()) / (df['width'].max()-df['width'].min()) 
df['height'] =(df['height']-df['height'].mean())/df['height'].std()
```

```python
#turning categorical values into numerical varaibles one hot vector 

dummy_variable=pd.get_dummies(df['fuel-type'])

dummy_variable.rename(columns={'diesel':'fuel-type-diesel','gas':'fuel-type-gas'},inplace=True)

df=pd.concat([df,dummy_variable],axis=1)

dummy_variable1=pd.get_dummies(df['aspiration'])

dummy_variable1.rename(columns={'std':'aspiration-std','turbo':'aspiration-turbo'},inplace=True)

df=pd.concat([df,dummy_variable1],axis=1)
```

```python
#remove fuel-type column

df.drop('fuel-type',axis=1,inplace =True)

df.drop('aspiration',axis=1,inplace=True)
```
