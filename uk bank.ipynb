##### COUNT ING OF GENDER

```python

df1=df[['Customer ID','Gender']].groupby(by='Gender',as_index=True).count()
df1=df1.reset_index()
plt.pyplot.bar(df1['Gender'].unique().tolist(),df1['Customer ID'])
```

##### count customer per job

```python
#count customer per job
df2=df[['Customer ID','Job Classification']].groupby(by='Job Classification',as_index=True).count()
#df2.reset_index(inplace=True)
plt.pyplot.pie(df2['Customer ID'] ,labels=df2.index,colors=['#4CAF99','cyan','blue'],autopct='%1.1f%%')
```

##### MAX OF BALANCE  & MIN OF BALANCE 

```python
print('Max of Balance'+':'+str(df['Balance'].min()),'Max of Balance'+':' +str(df['Balance'].max()))
```

##### EXTRACT DAY OF WEEK

```python
df['Date Joined']=pd.to_datetime(df['Date Joined'])
df['Day of week']=df['Date Joined'].dt.day_name()
```

##### EXTRACT MONTH 

```python
df['Date Joined']=pd.to_datetime(df['Date Joined'])
df['month']=df['Date Joined'].dt.month_name()
df
```

##### SUM BALANCE PER DAY

```python
df2=df[['Balance','Day of week']].groupby(by='Day of week').sum()
df2.reset_index(inplace=True)
plt.pyplot.plot(df2['Day of week'],df2['Balance'])
plt.pyplot.show()
```

##### SUM BALANCE PER MONTH

```python
plt.pyplot.figure (figsize= (12,5)) 
df3=df[['month','Balance']].groupby(by='month').sum()
df3.reset_index(inplace=True)
plt.pyplot.plot(df3['month'],df3['Balance'])
plt.pyplot.show()
```
